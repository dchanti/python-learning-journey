### while Loop 

```python
# Reverse a number

# Input: 123 → 321
num=123
new_num=0
while(num>0):
    new_num=(new_num*10)+(num%10)
    num//=10
print(new_num)    
```

### nested loops

```python
# Print pattern:

# *
# **
# ***
# ****

for i in range(1,5):
    for j in range(i):
        print("*",end=" ")
    print()    
```
