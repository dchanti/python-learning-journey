```python
# Check if a number is divisible by 5
```

```python
number=int(input("enter the number"));
remainder=number%10
if remainder==5 or remainder==0:
    print(f"{remainder} {number} is divisible by 5")
else :
    print(f"{remainder} {number} is not divisible by 5")    
```
