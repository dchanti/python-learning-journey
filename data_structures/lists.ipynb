```python
#slicibng
array=[1,2,"hello",True,1.2,1,2,3,4,5,6,7,78,8,99]
# slicing basic [start:end]  end is excluded 
print(array[0:])  ## from 0 index to last index 
print(array[3:])   ## from third index to last index 
print(array[:]) ## to print entire list

print(array[1:3]) 
print(array[3:8])
print(array[4:9])

## by default the above also has step 1 means increase +1 every time index 
## using step [start : end : step ] 
print(array[::2]) #it takes two two jumps starts from fiirst to end 
print(array[1::3])  ## here we have given start index so index 1 to 3 number jumps
print(array[::-1]) ## actually it prints entire list reverse it goes end to start by decreasing -1 
## python is very intellignet it does its things 
```

### some important methods we can perform on list 
like min max sum and sort reverese and index and count , in not in

```python
array=[1,2,45,6,7,34,2,3,45,6,777,8,8,8,909990,4,455635,65,365,6655,6]
minimum=min(array);
print(f"minimum numeber is = {minimum}")
maximum=max(array)
print(f"maximum numeber is = {maximum}")
##  find index of an element first occurence 
present_at=array.index(8)
print(f"the index of the element ={present_at}")

### count occurence of a a number 
kount=array.count(6)
print(f"count of the numnber= {kount}")


## sort sorts the list sorted return sorted list
# array.sort() this also works
sorted_array=list(sorted(array))
print(f"soted arry : {sorted_array}")



### can also reverse the array or use reveresed to get the reveresed  array

reveresed_array=list(reversed(array))
print(f"Revered array = {reveresed_array}")


## in , not in concept can be applkied on all 
# data structures to check if an element is presntn in another or not

print(3 in array)
if 12 not in array:
    print("heyyyyy")
```

### Zip combines ti lists and makes a tuple 
zip(list1, list2)

```python
list1=[1,2,3,4,7]
list2=["ravi","suresh","bhargav","prakah","Jp"]
print(list(zip(list1,list2)))
print(dict(zip(list2,list1)))
```

```python
### we can also unzip  * F
list4=[[1,2,3],
    [4,5,6],
    [7,8,9]]
a,b,c=zip(*list4)
print(a)
print(b)
print(c)
```

### 2d  lists 
creatinon access update and iterate 


```python
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
    ]

##access
print(matrix[0])
print(matrix[1][2])

##update simple
matrix[1][2]=25
print(matrix)

##iterate type 1 
for mat in matrix:
    print(mat)

## type 2
for sub in matrix:
    for su in sub:
        print(su, end =" ")
    print()    
```

### list comprehension thsi can work on any 

basic syntax [expresssion for item in iterable]

```python
## example 1 (square of numbers)
squares=[i**2 for i in range(5)]
print( squares)
```

```python
# example2: check each element and insert only even number (for if)
even_no=[x  for x in range(10) if x%2==0 ]
print(even_no)
```

```python
# example 3: using for if and else (if else for)
numk=[1,2,3,4,5,6,7,8,9]
labels= ["even" if x%2==0 else "odd" for x in numk]
print (labels)
```
