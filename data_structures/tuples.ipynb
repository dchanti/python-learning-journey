```python
### tuple constructor 
wor=tuple("hello Bro")
print(wor);

war=tuple(range(5))
print(war)
```

```python
array=(1,2,"hello",True,1.2,1,2,3,4,5,6,7,78,8,99)
# slicing basic [start:end]  end is excluded 
print(array[0:])  ## from 0 index to last index 
print(array[3:])   ## from third index to last index 
print(array[:]) ## to print entire list
print(array[:5]) ## from starting index to 5 th index 

print(array[1:3]) 
print(array[3:8])
print(array[4:9])

## by default the above also has step 1 means increase +1 every time index 
## using step [start : end : step ] 
print(array[::2]) #it takes two two jumps starts from fiirst to end 
print(array[1::3])  ## here we have given start index so index 1 to 3 number jumps
print(array[::-1])
```

```python
### concatenation of tuples
lpk1=(1,2,3,4)
lpk2=(3,4,5,6)
lpk3=lpk1+lpk2
print(lpk3)
```

#### IMportant Concept Packing and unpacking 

```python
# tuple can also be written in 
x= 1,2 ,"buffalo", 5  ## this is packing we are assigning multiple elements to a single varaibl;e 

## tuple unpacking 
y=(2,3)
a,b=y
print(a , b)

```

```python
### Unpacking with *rest operator

q=(1,2,3,4,5,67)
a,*b,c=q
print(a)
print(b)
print(c)
##
a,c,*b=q
print(a)
print(c)
print(b)
```

```python
### swapping inpython is simple 
a=5;b=6
a,b=b,a
print(a)
print(b)
```

```python
## iterationn
q=(1,2,3,4,5,67)
for k in q:
    print(k , end= " ")
```

```python
### Nested Tuples
matrix=(
    (1,2,3),
    (4,5,6),
    (7,8,9)
)
##access
print(matrix[0])
print(matrix[1][2])


##iterate type 1 
for mat in matrix:
    print(mat)

## type 2
for sub in matrix:
    for su in sub:
        print(su, end =" ")
    print()  
```

```python
# unpacking in a loop 
students=(("alice",85),
          ("bob",96),
          ("alex",78)
          )

for name,marks in students:
    print(f"{name} => {marks}")


    


```
